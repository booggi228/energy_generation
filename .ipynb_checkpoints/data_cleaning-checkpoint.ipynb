```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data.xlsx', header=2)
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# drop blank rows
df.dropna(subset=["Найменування суб'єкта госпощдарювання", "Область"], inplace=True)
```

```python
df.describe()
```

```python
# filling nan values
df['Зелений Тариф ЕЦ'] = df['Зелений Тариф ЕЦ'].fillna(df['Зелений Тариф ЕЦ'].mean())
df['Потужність, МВт'] = df['Потужність, МВт'].fillna(df['Потужність, МВт'].mean())
df['Тип'] = df['Тип'].fillna('мГЕС')
```

```python
# drop columns
df.drop(['Unnamed: 0', 'Вер.19', 'Жов.19', 'Лис.19', 'Гру.19', 'Дата введення в експлуатацію'], axis=1, inplace=True)
```

```python
months = df.iloc[:, 12:]

# replace nan values in month columns with 0
for cols in months:
    df[cols].replace(np.nan, 0, inplace=True)    
```

```python
df.info()
```

```python
df["Підтип"].value_counts()
```

```python
# rename features and deal with missing data
df['Квартал'] = df['Квартал'].replace('IV', 4).replace('I', 1).replace('III', 3).replace('II', 2)
df['Квартал'] = df['Квартал'].replace(5, 2)
df['Квартал'] = df['Квартал'].fillna(2)

df["% надбавки"] = df["% надбавки"].replace(np.nan, 0)
df["Місяць"] = df["Місяць"].fillna(4)
df["Підтип"] = df["Підтип"].replace(np.nan, 'unknown')
```

```python
# remove unimportant labels
df['Тип'] = df['Тип'].apply(lambda x: x.split('(')[0])
df['Підтип'] = df['Підтип'].apply(lambda x: x.split('(')[0])
```

```python
df.to_csv('excel_data.csv', index=False)
```
